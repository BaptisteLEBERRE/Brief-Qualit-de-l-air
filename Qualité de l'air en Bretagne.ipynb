## Préparations

Importer les modules nécessaires :

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Lire le document :

```python
data = pd.read_csv("ind_bretagne_agglo.csv")
```

Identifier les aires urbaines utilisées :

```python
set(data["lib_zone"])
```

Rajouter les coordonnées latitude et longitude :

```python
data["latitude"] = 0
data["longitude"] = 0
```

```python
data["latitude"][data["lib_zone"] == 'Brest Métropole'] = 48.390394
data["longitude"][data["lib_zone"] == 'Brest Métropole'] = -4.486076

data["latitude"][data["lib_zone"] == 'CA Golfe du Morbihan - Vannes'] = 47.658236
data["longitude"][data["lib_zone"] == 'CA Golfe du Morbihan - Vannes'] = -2.760847

data["latitude"][data["lib_zone"] == 'CA Lorient'] = 47.7482524
data["longitude"][data["lib_zone"] == 'CA Lorient'] = -3.3702449

data["latitude"][data["lib_zone"] == 'CA Quimper Bretagne Occidentale'] = 47.997542
data["longitude"][data["lib_zone"] == 'CA Quimper Bretagne Occidentale'] = -4.097899

data["latitude"][data["lib_zone"] == 'CA Saint-Brieuc Armor'] = 48.51418
data["longitude"][data["lib_zone"] == 'CA Saint-Brieuc Armor'] = -2.765835

data["latitude"][data["lib_zone"] == 'CA du Pays de Saint-Malo (Saint-Malo Agglomération)'] = 48.649337
data["longitude"][data["lib_zone"] == 'CA du Pays de Saint-Malo (Saint-Malo Agglomération)'] = -2.025674

data["latitude"][data["lib_zone"] == 'Rennes Métropole'] = 48.117266
data["longitude"][data["lib_zone"] == 'Rennes Métropole'] = -1.6777926
```

La Dataframe retravaillée :

```python
data
```

```python
data.dtypes
```

Séparer la Dataframe suivant les dates de confiment et non confinement :

```python
data1 = data[data["date_ech"] < "2020-03-17T10:00:00"]
data2 = data[data["date_ech"] > "2020-03-16T10:00:00"][data["date_ech"] < "2020-05-12T10:00:00"]
data3 = data[data["date_ech"] > "2020-05-11T10:00:00"][data["date_ech"] < "2020-10-30T10:00:00"]
data4 = data[data["date_ech"] > "2020-10-29T10:00:00"][data["date_ech"] < "2020-12-16T10:00:00"]
```

Appliquer la méthode groupby pour écraser les Dataframe suivant leurs moyennes :

```python
data1 = data1.groupby("lib_zone").mean()
data2 = data2.groupby("lib_zone").mean()
data3 = data3.groupby("lib_zone").mean()
data4 = data4.groupby("lib_zone").mean()
```

Référencer les noms de villes pour les points, et les coordonnées pour l'image de fond :

```python
names = ["Brest", "Vannes", "Lorient", "Quimper", "Saint-Brieuc", "Saint-Malo", "Rennes"]
coord = [data["longitude"].min()-0.7, data["longitude"].max()+0.8, data["latitude"].min()-0.4, data["latitude"].max()+0.3]
```

## Les cartes

```python
plt.imshow(plt.imread("bretagne39.gif"), extent=coord)

scatter1 = plt.scatter(data1["longitude"], data1["latitude"], c=data1["valeur"], s=200, vmin=3, vmax=5)
scatter1

for i, txt in enumerate(data1["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data1["longitude"][i], data1["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 1 janvier-16 mars 2020")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar1 = fig.colorbar(scatter1, ticks=[3, 4, 5])
cbar1.ax.set_yticklabels(["Bon", "Bon", "Moyen"])
plt.show()
```

```python
plt.imshow(plt.imread("bretagne39.gif"), extent=coord)

scatter2 = plt.scatter(data2["longitude"], data2["latitude"], c=data2["valeur"], s=200, vmin=3, vmax=5)
scatter2

for i, txt in enumerate(data2["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data2["longitude"][i], data2["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 17 mars-11 mai 2020 (confinement)")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar2 = fig.colorbar(scatter2, ticks=[3, 4, 5])
cbar2.ax.set_yticklabels(["Bon", "Bon", "Moyen"])
plt.show()
```

```python
plt.imshow(plt.imread("bretagne39.gif"), extent=coord)

scatter3 = plt.scatter(data3["longitude"], data3["latitude"], c=data3["valeur"], s=200, vmin=3, vmax=5)
scatter3

for i, txt in enumerate(data3["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data3["longitude"][i], data3["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 12 mai-29 octobre 2020")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar3 = fig.colorbar(scatter3, ticks=[3, 4, 5])
cbar3.ax.set_yticklabels(["Bon", "Bon", "Moyen"])
plt.show()
```

```python
plt.imshow(plt.imread("bretagne39.gif"), extent=coord)

scatter4 = plt.scatter(data4["longitude"], data4["latitude"], c=data4["valeur"], s=200, vmin=3, vmax=5)
scatter4

for i, txt in enumerate(data4["valeur"]):
    plt.annotate((names[i], round(txt, 2)), (data4["longitude"][i], data4["latitude"][i]))

plt.title("Moyennes de l'indice de pollution, 30 octobre-15 décembre 2020 (confinement)")
fig = plt.gcf()
fig.set_size_inches(15, 5)
cbar4 = fig.colorbar(scatter4, ticks=[3, 4, 5])
cbar4.ax.set_yticklabels(["Bon", "Bon", "Moyen"])
plt.show()
```

## Observations

Le premier confinement de mars à mai 2020 a eu un impact négatif sur la qualité de l'air des villes bretonnes, avec plus de jours de qualité d'air proche de "Moyen" (5) et "Médiocre" (6) par rapport à une majorité de jours "Bon" (3-4). Le deuxième confinement d'octobre à décembre a au contraire présenté une très légère amélioration.

La qualité de l'air des villes bretonnes reste par ailleurs généralement "Bonnes".
